```python
def where_to_go(mood):
    if mood == "hungry":
        print ("Jollibee")
    elif mood == "bored":
        print ("Park")
    else:
        print ("Stay at home")

print(where_to_go("hungry"))
```

```python
print(where_to_go("bored"))
```

```python
def drink(need):
    if need == "to stay awake":
        print ("Coffee")
    elif need == "to stay energize":
        print ("Redbull")
    elif need == "to fall asleep":
        print ("Milk")
    else:
        print ("Water")

print (drink("to stay energize"))
```

```python
print(drink("to stay awake"))
```

```python
print(drink("to fall asleep"))
```
